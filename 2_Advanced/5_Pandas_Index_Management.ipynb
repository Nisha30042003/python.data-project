```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#new data frame that filters data only for jobs in india
df_india=df[df['job_country'] == 'India']
```

```python
df_india.head()
```

```python
#to give aproper indexing to the jobs above
df_india.reset_index(inplace=True)
df_india
```

```python
#to drop the indexing we did before
df_india.set_index('job_index', inplace=True)

df_india
```

```python
#to short the dataframes with lable
median_pivot=df_india.pivot_table(values='salary_year_avg',index='job_title_short', aggfunc=['median', 'min', 'max'])
```
